```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
a=pd.read_csv('titanic-passengers.csv',sep=';')
a['Cabin'].str.encode('utf-8')
#a.describe()
#a.isnull().sum()
#a.info()
#a.head()
a["Age"].fillna(int(a["Age"].mean()),inplace=True)
a['Cabin'] = a['Cabin'].apply(lambda v: np.random.choice(["C23","C25","C27","B96","B98"]))
a['Embarked'] = a['Embarked'].apply(lambda p: np.random.choice(["S","C","Q"]))
#a.isnull().sum()
a=a.set_index(["PassengerId"])
a=a.sort_index()
#the DataFrame is now all set up time to start the visualization!
#agecount=sns.countplot(x='Sex',data=a)
c=sns.FacetGrid(a,col="Survived")
c.map(plt.hist,"Sex")
d=sns.FacetGrid(a,col="Survived")
d.map(plt.hist,"Age")
b=sns.FacetGrid(a,row="Sex",col="Survived")
b.map(sns.barplot,"Age")
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
#plot_correlation_map(a)
b=pd.read_csv('titanic-passengers.csv',sep=';')
#we created another data frame out of the same csv file to use it values so we dont have to use some random values (the added ones with the fillna)
b=b.dropna()
Survival={"Survived":{"Yes":1, "No": 0}} 
b.replace(Survival, inplace=True)
embarked=b[["Embarked", "Survived"]].groupby(["Embarked"], as_index=True).mean()
parch=b[["Parch", "Survived"]].groupby(["Parch"], as_index=True).mean()
parch.max(axis=None, skipna=None, level=None, numeric_only=None)
parch.idxmax()
parch
embarked.max(axis=None, skipna=None, level=None, numeric_only=None)
embarked.idxmax()
embarked
#we realise from these analysisations the importance of the age and sex in the survival of the individual
#we realise that the members with parch 3 have the largest number of survivals
#we realise that the passengers which where in the Embarked C are the most ones who survived
plot_correlation_map(b)
#the plot_correlation_map is just a way to show statistics using the colors from the warmer to the less warmer
pclass=b[["Pclass", "Survived"]].groupby(["Pclass"], as_index=True).mean()
pclass
```
